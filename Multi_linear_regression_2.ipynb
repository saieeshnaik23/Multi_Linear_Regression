```python
adms_mkts = smf.ols('Profit~ADMS+MKTS',data = startup1).fit()
adms_mkts.tvalues,adms_mkts.pvalues
```

```python
coll_rds = smf.ols('RDS~ADMS+MKTS',data = startup1).fit().rsquared
vif_rds = 1/(1 - coll_rds)

coll_adms = smf.ols('ADMS~RDS+MKTS',data = startup1).fit().rsquared
vif_adms = 1/(1-coll_adms)

coll_mkts = smf.ols('MKTS~ADMS+RDS',data = startup1).fit().rsquared
vif_mkts = 1/(1-coll_mkts)

d1 = {'varaibles':['RDS','ADMS','MKTS'],'vif':[vif_rds,vif_adms,vif_mkts]}
df = pd.DataFrame(d1)
df
```
